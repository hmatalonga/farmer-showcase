#### Context for some of the columns:

* **fk** is simply the foreign key used to joined all tables
* **timestamp** is the actual time when the row was recorded on the Android device
* **screen_brightness** are values between 0 and 255 or -1 meaning that is automatically set by the OS
* **current_average** is the battery current average in microAmperes
* **up_time and sleep_time** are in seconds
* **wifi_link_speed** is Link speed in Mbps
* **mcc** is Numeric country code
* **mnc** is Numeric network code
